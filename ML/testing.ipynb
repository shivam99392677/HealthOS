````python
import json
from datetime import datetime
import re

def clean_and_format_fhir_response(response_data):
    """
    Clean and format the FHIR mapping response into a clear JSON structure
    """
    
    # If response_data is a string, parse it as JSON
    if isinstance(response_data, str):
        try:
            # Extract JSON if it's embedded in text
            json_match = re.search(r'```json\n(.*?)\n```', response_data, re.DOTALL)
            if json_match:
                response_data = json.loads(json_match.group(1))
            else:
                response_data = json.loads(response_data)
        except json.JSONDecodeError:
            return {"error": "Invalid JSON format"}
    
    # Create formatted output structure
    formatted_output = {
        "summary": response_data.get("summary", ""),
        "fhir_mapping": {
            "patient": format_patient_info(response_data.get("patient", {})),
            "encounter": format_encounter_info(response_data.get("encounter", {})),
            "conditions": format_conditions(response_data.get("conditions", [])),
            "observations": format_observations(response_data.get("observations", [])),
            "vital_signs": extract_vital_signs(response_data.get("observations", []))
        },
        "metadata": {
            "generated_at": datetime.now().isoformat(),
            "version": "FHIR R4",
            "total_conditions": len(response_data.get("conditions", [])),
            "total_observations": len(response_data.get("observations", []))
        }
    }
    
    return formatted_output

def format_patient_info(patient_data):
    """Format patient information"""
    if not patient_data:
        return {}
    
    name = patient_data.get("name", [{}])[0]
    
    return {
        "full_name": f"{name.get('given', [''])[0]} {name.get('family', '')}".strip(),
        "first_name": name.get('given', [''])[0] if name.get('given') else "",
        "last_name": name.get('family', ''),
        "gender": patient_data.get("gender", "").capitalize(),
        "birth_date": patient_data.get("birthDate", ""),
        "age": calculate_age(patient_data.get("birthDate", ""))
    }

def format_encounter_info(encounter_data):
    """Format encounter information"""
    if not encounter_data:
        return {}
    
    reasons = encounter_data.get("reason", [])
    primary_reason = reasons[0] if reasons else {}
    
    return {
        "visit_reason": primary_reason.get("text", ""),
        "visit_date": datetime.now().strftime("%Y-%m-%d"),
        "visit_type": "Initial Consultation"
    }

def format_conditions(conditions_list):
    """Format conditions with detailed information"""
    formatted_conditions = []
    
    for condition in conditions_list:
        formatted_cond = {
            "description": condition.get("text", ""),
            "clinical_status": condition.get("status", "active").capitalize(),
            "code": {
                "value": condition.get("code", ""),
                "system": condition.get("system", ""),
                "display": condition.get("display", "")
            } if condition.get("code") else None
        }
        formatted_conditions.append(formatted_cond)
    
    return formatted_conditions

def format_observations(observations_list):
    """Format observations with detailed information"""
    formatted_observations = []
    
    for obs in observations_list:
        formatted_obs = {
            "description": obs.get("text", obs.get("display", "")),
            "category": categorize_observation(obs),
            "value": obs.get("value"),
            "unit": obs.get("unit", ""),
            "code": {
                "value": obs.get("code", ""),
                "system": obs.get("system", ""),
                "display": obs.get("display", "")
            } if obs.get("code") else None
        }
        formatted_observations.append(formatted_obs)
    
    return formatted_observations

def extract_vital_signs(observations_list):
    """Extract vital signs specifically"""
    vital_signs = {}
    
    for obs in observations_list:
        if obs.get("code") == "8480-6":  # Systolic BP
            vital_signs["blood_pressure_systolic"] = f"{obs.get('value')} {obs.get('unit', 'mmHg')}"
        elif obs.get("code") == "8462-4":  # Diastolic BP
            vital_signs["blood_pressure_diastolic"] = f"{obs.get('value')} {obs.get('unit', 'mmHg')}"
        elif obs.get("code") == "72593-5":  # Pain scale
            vital_signs["pain_level"] = f"{obs.get('value')}/10"
    
    if "blood_pressure_systolic" in vital_signs and "blood_pressure_diastolic" in vital_signs:
        vital_signs["blood_pressure"] = f"{vital_signs['blood_pressure_systolic']}/{vital_signs['blood_pressure_diastolic']}"
    
    return vital_signs

def categorize_observation(observation):
    """Categorize observation type"""
    text = observation.get("text", "").lower() or observation.get("display", "").lower()
    
    if any(word in text for word in ["blood pressure", "systolic", "diastolic"]):
        return "Vital Sign"
    elif any(word in text for word in ["pain", "headache"]):
        return "Symptom"
    elif any(word in text for word in ["coffee", "water", "diet", "exercise"]):
        return "Lifestyle"
    elif any(word in text for word in ["sleep"]):
        return "Sleep Pattern"
    elif any(word in text for word in ["muscle", "tightness", "spasm"]):
        return "Physical Finding"
    else:
        return "General Observation"

def calculate_age(birth_date):
    """Calculate age from birth date"""
    if not birth_date:
        return "Unknown"
    
    try:
        birth = datetime.strptime(birth_date, "%Y-%m-%d")
        today = datetime.now()
        age = today.year - birth.year - ((today.month, today.day) < (birth.month, birth.day))
        return f"{age} years"
    except:
        return "Unknown"

def save_to_file(formatted_data, filename="fhir_mapping_output.json"):
    """Save formatted data to a JSON file"""
    with open(filename, 'w') as f:
        json.dump(formatted_data, f, indent=2, ensure_ascii=False)
    print(f"✅ Data saved to {filename}")

def print_pretty_json(data):
    """Print JSON in a pretty format"""
    print(json.dumps(data, indent=2, ensure_ascii=False))

# Example usage with your response
if __name__ == "__main__":
    # Your response data
    # response = {'response': '```json\n{\n  "summary": "Patient Michael Chen, 42, presents with a few months history of intense, daily tension-type headaches, rated 6-7/10, primarily at temples and spreading to neck. Aggravated by screen time, poor sleep, skipped meals, and associated with teeth grinding and high stress. Physical exam revealed muscle tension in trapezius and jaw clicking. The doctor recommended lifestyle modifications including ergonomic changes, reduced caffeine, increased water intake, regular exercise, a night guard for bruxism, stress management techniques, and improved sleep hygiene.",\n  "patient": {\n    "name": [\n      {\n        "text": "Michael Chen",\n        "family": "Chen",\n        "given": [\n          "Michael"\n        ]\n      }\n    ],\n    "gender": "male",\n    "birthDate": "1982-06-15"\n  },\n  "encounter": {\n    "reason": [\n      {\n        "text": "intense headaches",\n        "code": "",\n        "system": "",\n        "display": ""\n      }\n    ]\n  },\n  "conditions": [\n    {\n      "text": "intense headaches",\n      "code": "25064002",\n      "system": "http://snomed.info/sct",\n      "display": "Headache",\n      "status": "active"\n    },\n    {\n      "text": "tension-type headaches",\n      "code": "274636001",\n      "system": "http://snomed.info/sct",\n      "display": "Tension-type headache",\n      "status": "active"\n    },\n    {\n      "text": "Light bothers me a bit",\n      "code": "24826004",\n      "system": "http://snomed.info/sct",\n      "display": "Photophobia",\n      "status": "active"\n    },\n    {\n      "text": "grinding my teeth at night",\n      "code": "194017009",\n      "system": "http://snomed.info/sct",\n      "display": "Bruxism",\n      "status": "active"\n    },\n    {\n      "text": "jaw sometimes feels sore in the morning",\n      "code": "27464000",\n      "system": "http://snomed.info/sct",\n      "display": "Jaw pain",\n      "status": "active"\n    },\n    {\n      "text": "It clicks a little on the left side (jaw)",\n      "code": "30177001",\n      "system": "http://snomed.info/sct",\n      "display": "Clicking jaw",\n      "status": "active"\n    },\n    {\n      "text": "Irritable with my wife and kids",\n      "code": "128485002",\n      "system": "http://snomed.info/sct",\n      "display": "Irritability",\n      "status": "active"\n    },\n    {\n      "text": "stress levels are pretty high right now",\n      "code": "735950004",\n      "system": "http://snomed.info/sct",\n      "display": "Stress",\n      "status": "active"\n    }\n  ],\n  "observations": [\n    {\n      "code": "8480-6",\n      "system": "http://loinc.org",\n      "display": "Systolic blood pressure",\n      "value": 118,\n      "unit": "mmHg"\n    },\n    {\n      "code": "8462-4",\n      "system": "http://loinc.org",\n      "display": "Diastolic blood pressure",\n      "value": 76,\n      "unit": "mmHg"\n    },\n    {\n      "code": "72593-5",\n      "system": "http://loinc.org",\n      "display": "Pain severity - 0-10 rating",\n      "value": 7,\n      "unit": "/10"\n    },\n    {\n      "text": "significant tightness in your trapezius muscles",\n      "code": "24920005",\n      "system": "http://snomed.info/sct",\n      "display": "Muscle spasm"\n    },\n    {\n      "text": "A little tightness on the right side (neck) when I turn all the way",\n      "code": "24920005",\n      "system": "http://snomed.info/sct",\n      "display": "Muscle spasm"\n    },\n    {\n      "text": "4 or 5 cups of coffee daily",\n      "code": "160654006",\n      "system": "http://snomed.info/sct",\n      "display": "Coffee consumption"\n    },\n    {\n      "text": "2 glasses of water a day",\n      "code": "364640000",\n      "system": "http://snomed.info/sct",\n      "display": "Water intake"\n    },\n    {\n      "text": "sits for 3 or 4 hours straight at computer",\n      "code": "224978000",\n      "system": "http://snomed.info/sct",\n      "display": "Sedentary lifestyle"\n    },\n    {\n      "text": "fallen off that habit of exercise",\n      "code": "229712009",\n      "system": "http://snomed.info/sct",\n      "display": "Lack of exercise"\n    },\n    {\n      "text": "snacks a lot on chips, cookies",\n      "code": "225577004",\n      "system": "http://snomed.info/sct",\n      "display": "Snack eating habit"\n    },\n    {\n      "text": "doesn\'t sleep well",\n      "code": "398910005",\n      "system": "http://snomed.info/sct",\n      "display": "Sleep problem"\n    },\n    {\n      "text": "watches TV in bed until falls asleep",\n      "code": "442083002",\n      "system": "http://snomed.info/sct",\n      "display": "Poor sleep hygiene"\n    }\n  ]\n}\n```'}
    
    # Extract the actual JSON from the response
    if isinstance(response, dict) and 'response' in response:
        json_str = response['response']
        # Clean and format
        formatted_result = clean_and_format_fhir_response(json_str)
        
        # Print pretty JSON
        print("📋 FORMATTED FHIR MAPPING OUTPUT")
        print("=" * 50)
        print_pretty_json(formatted_result)
        
        # Save to file
        save_to_file(formatted_result)
        
        # Also create a simplified summary
        print("\n📊 QUICK SUMMARY")
        print("=" * 50)
        patient = formatted_result['fhir_mapping']['patient']
        print(f"Patient: {patient['full_name']} ({patient['age']}, {patient['gender']})")
        print(f"Visit Reason: {formatted_result['fhir_mapping']['encounter']['visit_reason']}")
        print(f"Conditions: {len(formatted_result['fhir_mapping']['conditions'])} identified")
        print(f"Vitals: {formatted_result['fhir_mapping']['vital_signs'].get('blood_pressure', 'N/A')}")
        print(f"Pain Level: {formatted_result['fhir_mapping']['vital_signs'].get('pain_level', 'N/A')}")
````
